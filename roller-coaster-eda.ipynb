```python
df.loc[df.duplicated(subset=['Coaster_Name'])]
```

```python
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

We can see from above output that there are some roller coasters mentioned twice or more, that shouldn't be happening because a roller coaster should only be opened once. Maybe there is an error, etc. So, we should handle this issue. Notice that all values are identical except for the 'Year_Introduced', maybe the roller coaster is offline and then online again.

```python
df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date_Clean']).sum()
```

```python
df.loc[df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date_Clean'])]
```

```python
df.query('Coaster_Name == "Crystal Beach Cyclone" & Location == "Crystal Beach Park" & Opening_Date_Clean == "1926-01-01"')
```

```python
# remove entries yang coastername, location, openingdate nya sama persis
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date_Clean'])].reset_index(drop=True)

# reset index buat ngereset indexnya semua biar gaada yang lompat2
# drop true biar kolom 'index' ilang
```

# **5 Ask Questions About Data**
-  What are the locations with the fastest roller coasters with minimum 10 coasters in those locations?

```python
df.head()
```

```python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_Mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', title='Top Coaster Locations by Average Speed')

ax.set_xlabel('Average Speed')
ax.set_ylabel('Location')
plt.show()
```
